```python
from typing import Iterable
def perm(remain:Iterable, n:int, cur:list=None):
    if None == cur:
        cur = []
    if n == 0:
        yield cur
        return
    if not remain:
        return
    for i, r in enumerate(remain):
        for x in perm(remain[:i] + remain[i+1:], n-1, cur + [remain[i],]):
            yield x

for x in perm('abcd', 2):
    print(x)


```

```python
from typing import Iterable

def comb(remain:Iterable, n:int, cur:list=None):
    if cur == None:
        cur = []
    if n == 0:
        yield cur
        return
    if not remain:
        return
    c = remain[0]
    for x in comb(remain[1:], n-1, cur + [c]):
        yield x
    for x in comb(remain[1:], n, cur):
        yield x
    return

for x in comb('abcd', 2):
    print(x)
```

```python
# 2-sum
# find two distinct numbers in the array that add up to the target value
def sum_2(data:list[int], s:int) -> bool:
    if len(data) < 2:
        return False
    rec = {}
    for n in data:
        if s - n in rec:
            return True
        rec[s-n] = True
    return False
```

```python
# max sub-array sum
# Given an array arr[], find the subarray that has the maximum sum and return its sum.
# dp[i] = max(dp[i-1] + data[i], data[i])

def max_subarr_sum(data:list[int]) -> int:
    if not data:
        return 0
    N = len(data)
    cur_max = cur = data[0]
    for n in data[1:]:
        cur = max(cur + n, n)
        cur_max = max(cur, cur_max)
    return cur_max

def test():
    test_data = [
        ([2, 3, -8, 7, -1, 2, 3], 11),
        ([-2, -4], -2),
        ([5, 4, 1, 7, 8], 25),
        ([], 0),
        ([-3], -3),
    ]
    for data, e in test_data:
        res = max_subarr_sum(data)
        print(f"{data} -> {res}.  {'Pass' if res==e else 'Fail'}" )

test()
```

```python
"""
Minimum in a Sorted and Rotated Array
e.g.  {5, 6, 1, 2, 3, 4},   Output: 1
The key is:  if data[mid] > data[high], low=mid+1 (go right)
   5, 6, 7, 8, 1, 2, 3, 4
              | 
   l  m     h                      -> data[mid] < data[high]
               l  m  h             -> data[mid] < data[high]
      l     m     h                -> data[mid] > data[high]
            l  m     h             -> data[mid] < data[high]
And the mid could be at the boarder of 2 sections. special check & match
            
"""

def srarray_find_min(data:list[int]) -> int:
    if not data:
        return None
    low, high = 0, len(data) - 1
    while low < high:
        if data[low] <= data[high]:
            return data[low]
        mid = (low + high) // 2
        if data[mid] > data[high]:
            low = mid + 1
        else:
            high = mid
    return data[low]

# logic complicated
def srarray_find_min1(data:list[int]) -> int:
    if not data:
        return None
    N = len(data)
    if N < 2:
        return data[0]
    low, mid, high = 0, 0, N-1
    while low <= high:
        mid = (low + high) // 2
        if mid < N-1 and data[mid] > data[mid+1]:
            return data[mid+1]
        if mid > 0 and data[mid] < data[mid-1]:
            return data[mid]
        if data[mid] == data[high]:
            return data[mid]
        elif data[mid] > data[high]:
            low = mid + 1
        else:
            high = mid - 1
    return data[mid]

def test():
    data = [
        (([],), None),
        (([3],), 3),
        (([4, 5, 6, 7, 0, 1, 2], ), 0),
        (([50, 10, 20, 30, 40],),10),
        (([3,3,3,3,3],),3),
        (([5,0,1,2,3],),0),
        (([0,1,2,3,4,5],),0),
        (([5,6,7,8,0],),0),
        (([5,6,7,8,0,1],),0),
        (([6,7,8,0,1,2],),0),        # min right to mid
        (([6,7,8,0,1,2,3],),0),      # min land on mid
    ]
    for d, e in data:
        result = srarray_find_min(*d)
        print(f"{d} -> {result}, {e}. {'pass' if result==e else 'fail'}")

test()
```

```python
"""
Search in a sorted and rotated Array
eg. arr[] = {4, 5, 6, 7, 0, 1, 2}, key = 0, -> index=4
                       |     
             l  m    h 
                         l    h
                   l     m     h
                l  m     h         

"""

def search_sr_array(data:list[int], v:int) -> int:
    if not data:
        return -1
    low, high = 0, len(data) - 1
    while low <= high:
        mid = (low + high) // 2
        if data[mid] == v:
            return mid
        if data[mid] < data[low]:
            if v > data[mid]:
                low = mid + 1
            else:
                high = mid - 1
                low += 1
        elif data[mid] > data[high]:
            if v < data[mid] and v >= data[low]:
                high = mid - 1
            elif v < data[low]:
                low = mid + 1
            else:
                low = mid + 1
                high -= 1
        else:
            if v < data[mid]:
                high = mid - 1
            else:
                low = mid + 1
    return -1

def test():
    data = [
        (([], 1),-1),
        (([3],3),0),
        (([3],4),-1),
        (([4, 5, 6, 7, 0, 1, 2],0),4),
        (([4, 5, 6, 7, 0, 1, 2],3),-1),
        (([4, 5, 6, 7, 0, 1, 2],5),1),
        (([4, 5, 6, 7, 0, 1, 2],1),5),
        (([4, 5, 6, 7, 0, 1, 2],4),0),
        (([4, 5, 6, 7, 0, 1, 2],2),6),
        (([50, 10, 20, 30, 40],10),1),
    ]
    for (d, v), e in data:
        result = search_sr_array(d, v)
        print(f"{d},{v} -> {result}, {e}. {'pass' if result==e else 'fail'}")

test()
```

```python
# merge two sorted arrays in place.  
#   arr1 size m+n, arr2 size n.  -> merge arr2 into arr1
# e.g.  arr1 [2,7,None,None,None,10], arr2 [5,8,12,14]

def merge_sorted_arr(a:list[int], b:list[int]) -> list[int]:
    loc = len(a) - 1
    for i in range(len(a)-1, -1, -1):
        if a[i] == None:
            continue
        a[loc] = a[i]
        loc -= 1
    for i in range(loc+1):
        a[i] = None
    p1, p2 = len(b), 0
    loc = 0
    # print(a, p1, p2)
    while p1 < len(a) or p2 < len(b):
        if p1 >= len(a):
            a[loc] = b[p2]
            p2 += 1
        elif p2 >= len(b):
            a[loc] = a[p1]
            p1 += 1
        else:
            if a[p1] < b[p2]:
                a[loc] = a[p1]
                p1 += 1
            else:
                a[loc] = b[p2]
                p2 += 1
        loc += 1
    return a


def test():
    data = [
        (([2, None, 7, None, None, 10, None], [5,8,12,14]), [2,5,7,8,10,12,14]), 
        (([2, 7, None, None, None, 10, None], [5,8,12,14]), [2,5,7,8,10,12,14]), 
        (([2, 7, 10, None, None, None, None], [5,8,12,14]), [2,5,7,8,10,12,14]), 
        (([None, None, None, None, 2, 7, 10], [5,8,12,14]), [2,5,7,8,10,12,14]), 
    ]    
    for (a,b),e in data:
        a2 = a.copy()
        result = merge_sorted_arr(a,b)
        print(f"{a2},{b} -> {result}. {'pass' if e == result else 'fail'}")

test()
```

```python
# Maximum sum such that no two elements are adjacent / Stickler Thief Editorial
# e.g   Input: hval[] = {5, 5, 10, 100, 10, 5}, Output: 110
#       Input: hval[] = {3, 2, 5, 10, 7},  Output: 15

def max_val(hval:list[int]) -> int:
    N = len(hval)
    if not hval:
        return 0
    if N == 1:
        return hval[0]
    dp = [0] * (N+1)    # dp[i] is the max value at i
    dp[0], dp[1] = hval[0], hval[1]
    for i in range(2, N):
        dp[i] = max(dp[i-2] + hval[i], dp[i-1])
    return dp[N-1]

def test():
    data = [
        ([], 0),
        ([3], 3),
        ([5, 5, 10, 100, 10, 5], 110),
        ([3, 2, 5, 10, 7], 15),
        ([3,3,3,3,3,], 9),
    ]
    for hval, e in data:
        result = max_val(hval)
        print(f"{hval} -> {result}, {'pass' if result == e else 'fail'} ")
    
test()
```

```python
"""
Sort elements by frequency
Print the elements of an array in the decreasing frequency 
if 2 numbers have the same frequency then print the one which came first

Examples:  
Input:  arr[] = {2, 5, 2, 8, 5, 6, 8, 8}
Output: arr[] = {8, 8, 8, 2, 2, 5, 5, 6}

Input: arr[] = {2, 5, 2, 6, -1, 9999999, 5, 8, 8, 8}
Output: arr[] = {8, 8, 8, 2, 2, 5, 5, 6, -1, 9999999}
"""
from collections import Counter

def sort_by_freq(arr:list[int]) -> list[int]:
    c = Counter(arr)
    sc = sorted(c, key=lambda x:c[x], reverse=True)
    output = []
    for k in sc:
        output.extend([k]*c[k])
    return output

def test():
    data = [
        ([],  []),
        ([3], [3]),
        ([5, 5, 10, 100, 10, 5], [5,5,5,10,10,100]),
        ([3, 2, 5, 10, 7], [3,2,5,10,7]),
        ([3,3,3,3,3,], [3,3,3,3,3]),
        ([2, 5, 2, 8, 5, 6, 8, 8], [8, 8, 8, 2, 2, 5, 5, 6]),
        ([2, 5, 2, 6, -1, 9999999, 5, 8, 8, 8], [8, 8, 8, 2, 2, 5, 5, 6, -1, 9999999]),
        ([2,2,2,1,3,3,3,4,4], [2,2,2,3,3,3,4,4,1]),
    ]
    for hval, e in data:
        result = sort_by_freq(hval)
        print(f"{hval} -> {result}, {'pass' if result == e else 'fail'} ")
    
test()
```

```python
"""
LLNode
LList
"""

class LLNode():
    def __init__(self, val) -> None:
        self.val = val
        self.next = None

class LList():
    def __init__(self) -> None:
        self.head = None
    
    @staticmethod
    def generate_llist(data:list[int]) -> LLNode:
        if not data:
            return None
        head = cur = None
        for d in data:
            node = LLNode(d)
            if not head:
                head = cur = node
                continue
            cur.next = node
            cur = node
        return head
    
    def generate(self, data:list[int]):
        self.head = self.generate_llist(data)

    @staticmethod
    def reverse_llist(head:LLNode) -> LLNode:
        if not head or not head.next:
            return head
        pre, cur = head, head.next
        while cur:
            nxt = cur.next
            cur.next = pre
            if pre == head:
                pre.next = None
            pre = cur
            cur = nxt
        return pre

    def reverse(self):
        self.head = self.reverse_llist(self.head)
    
    def __repr__(self) -> str:
        s = ''
        cur = self.head
        while cur:
            s += f'-> {cur.val}'
            cur = cur.next
        return s

def test():
    data = [i for i in range(10,0,-1)]
    ll = LList()
    ll.generate(data)
    print(ll)

    ll.reverse()
    print(ll)

    nl = ll.reverse_llist(ll.head)
    show_ll(nl)

test()
```

```python
"""
DLL  Double Linked List
"""
class DLLNode():
    def __init__(self, val:int) -> None:
        self.val = val
        self.pre = None
        self.next = None


def generate_dllist(data:list[int]) -> DLLNode:
    head = cur = None
    for d in data:
        node = DLLNode(d)
        if not head:
            head = node
            cur = node
        else:
            cur.next = node
            node.pre = cur
            cur = cur.next
    return head

def show_dll(head:DLLNode):
    cur = head
    while cur:
        print(f"->{cur.val}", end='')
        cur = cur.next
    print('')

def reverse_dll(head:DLLNode) -> DLLNode:
    pre = None
    cur = head
    new_head = None
    while cur:
        nxt = cur.next
        cur.next = pre
        cur.pre = nxt
        pre = cur
        cur = nxt
    new_head = pre
    new_head.pre = None
    return new_head


def test():
    dll = generate_dllist([i for i in range(10)])
    show_dll(dll)

    rdll = reverse_dll(dll)
    show_dll(rdll)


test()
```

```python
"""
reverse every group of k nodes in a single linked list. 
e.g
input: head: 1 -> 2 -> 3 -> 4 -> 5 -> 6 -> NULL, k = 2 
Output: head: 2 -> 1 -> 4 -> 3 -> 6 -> 5 -> NULL 


"""

def reverse_ll(head:LLNode) -> tuple[LLNode, LLNode]:
    if not head or not head.next:
        return head, None
    pre, cur = head, head.next
    tail = head
    ch = None
    while cur:
        nxt = cur.next
        cur.next = pre
        if pre == head:
            pre.next = None
        ch = cur
        pre = cur
        cur = nxt
    return ch, tail


def reverse_every_k(head:LLNode, k:int) -> LLNode:
    count = 0
    cur = head
    ch = ct = nh = cur    # cur_head, cur_tail, next_head
    dummy = LLNode(None)
    tail = dummy
    while cur:
        while cur and count < k - 1:
            cur = cur.next
            count += 1
        if cur:
            nh = cur.next
            cur.next = None
        else:
            nh = None
        show_ll(ch)
        tail.next, ct = reverse_ll(ch)
        tail = ct
        cur = ch = nh
        count = 0
    return dummy.next


def test_reverse_ll():
    data = [1,2,3,4,5,6,7]   
    ll = create_ll(data)
    show_ll(ll)
    rl, tail = reverse_ll(ll)
    show_ll(rl)
    print(tail.val)

def test():
    data = [1,2,3,4,5,6,7]   
    ll = create_ll(data)
    show_ll(ll)
    rl = reverse_every_k(ll, 2)
    show_ll(rl)

test()
```

```python
"""
rotate the linked list to the left by k places.
"""

def rotate_left_k(head:LLNode, k:int) -> LLNode:
    if not head or not head.next:
        return head
    N = 0
    cur = head
    tail = None
    while cur:
        N += 1
        tail = cur
        cur = cur.next
    k = k % N
    if not k:
        return head
    cur = head
    pre = None
    for i in range(k):
        pre = cur
        cur = cur.next
    nh = cur
    pre.next = None
    tail.next = head
    return nh

def test():
    ll = LList.generate_llist([1,2,3,4,5,6,7])
    show_ll(rotate_left_k(ll, 2))
    ll = LList.generate_llist([1,2,3,4,5,6,7])
    show_ll(rotate_left_k(ll, 6))
    ll = LList.generate_llist([1,2,3,4,5,6,7])
    show_ll(rotate_left_k(ll, 10))
    ll = LList.generate_llist([])
    show_ll(rotate_left_k(ll, 10))
    ll = LList.generate_llist([1])
    show_ll(rotate_left_k(ll, 10))
    ll = LList.generate_llist([1,2])
    show_ll(rotate_left_k(ll, 10))
    ll = LList.generate_llist([1,2])
    show_ll(rotate_left_k(ll, 9))



test()
```

```python
"""
Split a Circular Linked List into two halves (2 circular linked lists)

"""

def split_cll(head:LLNode) -> tuple[LLNode, LLNode]:
    if not head or head.next == head:
        return (head, None)   
    slow = fast = head
    pre_slow = pre_fast = None
    while True:
        pre_slow = slow
        slow = slow.next
        pre_fast = fast
        fast = fast.next.next
        if fast == head or fast.next == head:
            break
    h2 = slow
    pre_slow.next = head
    t2 = pre_fast
    while t2.next != head:
        t2 = t2.next
    t2.next = h2
    print(head.val, pre_slow.val, h2.val, t2.val)
    return head, h2

def test_cll(head:LLNode):
    tail, cur = None, head
    while cur:
        tail = cur
        cur = cur.next
    if tail:
        tail.next = head
    show_cll(head)

    h1, h2 = split_cll(head)
    show_cll(h1)
    show_cll(h2)

def test():
    head = LList.generate_llist([i for i in range(10)])
    test_cll(head)
    head = LList.generate_llist([i for i in range(9)])
    test_cll(head)
    head = LList.generate_llist([])
    test_cll(head)
    head = LList.generate_llist([1])
    test_cll(head)

test()


```

```python
# BTree

from collections import deque, defaultdict

class BTNode():
    def __init__(self, val) -> None:
        self.val = val
        self.left = None
        self.right = None

def generate_btree(data:list[int]) -> BTNode:
    if not data:
        return None
    root = BTNode(data[0])
    data = deque(data[1:])
    dq = deque([root,])
    while dq:
        cur = dq.popleft()
        if not data:
            cur.left = cur.right = None
            break
        ln = BTNode(data.popleft())
        cur.left = ln
        dq.append(ln)
        if not data:
            cur.right = None
            break
        rn = BTNode(data.popleft())
        cur.right = rn
        dq.append(rn)
        # print(cur.val, cur.left.val, cur.right.val)
    return root

def bt_bfs(root:BTNode):
    dq = deque([root, ])
    while dq:
        node = dq.popleft()
        if node:
            yield node.val
        else:
            yield None
        if node.left:
            dq.append(node.left)
        if node.right:
            dq.append(node.right)

def bt_bfs_lvl(root:BTNode) -> dict:
    if not root:
        return {}
    info = defaultdict(list)
    dq = deque([(root, 1), ])
    while dq:
        node, lvl = dq.popleft()
        info[lvl].append(node.val)
        if node.left:
            dq.append((node.left, lvl + 1))
        if node.right:
            dq.append((node.right, lvl + 1))
    print(info)
    return info

def show_btree(root:BTNode):
    info = bt_bfs_lvl(root)
    cur_lvl = 0
    for lvl in sorted(info.keys()):
        if lvl != cur_lvl:
            print(f'\n[{lvl}]: ', end='')
            cur_lvl = lvl
        for n in info[lvl]:
            print(f"{n}  ", end="")
    print("")


def test():
    data = [i for i in range(1,10)]
    root = generate_btree(data)
    print([n for n in bt_bfs(root)])
    show_btree(root)

test()
```

```python
"""
A Boolean Matrix Question
Given a boolean matrix mat[M][N] of size M X N, modify it such that 
if a matrix cell mat[i][j] is 1 (or true) then make all the cells of ith row and jth column as 1. 

"""

# T(M*N), S(N+M)
def booleanMatrix1(matrix):
    R = len(matrix)
    C = len(matrix[0])
    rows = [0] * R
    cols = [0] * C
    for r in range(R):
        rows[r] = 1 if any(matrix[r]) else 0
    for c in range(C):
        cols[c] = 1 if any([matrix[r][c] for r in range(R)]) else 0
    for r in range(R):
        matrix[r] = [1]*C if rows[r] else cols

```

```python
"""
Maximum size square sub-matrix with all 1s

Given a binary matrix, find out the maximum size square sub-matrix with all 1s. 
For example, consider the below binary matrix. 


"""

def maxSquare(mat:list[list[int]]) -> int:
    R = len(mat)
    C = len(mat[0])
    dp = [[0] * C for _ in range(R)]
    maxs = 0
    for r in range(0,R):
        for c in range(0,C):
            if r == 0 or c == 0:
                dp[r][c] = mat[r][c]
            elif not mat[r][c]:
                dp[r][c] = 0
            else:
                nmx = min([dp[r-1][c-1], dp[r-1][c], dp[r][c-1]]) + 1
                dp[r][c] = nmx
            maxs = max(maxs, dp[r][c])
    return maxs
```

```python
"""
max sum of submatrix

"""

from typing import List
from sys import maxsize

class Solution:
    def maxSquare(self, n : int, m : int, mat : List[List[int]]) -> int:
        dp = [[-maxsize] * m for _ in range(n)]
        dp[0][0][1] = mat[0][0]
        maxs = -maxsize
        max_at = [[[-maxsize] * min(m,n) for c in range(m)] for r in range(n)]
        for r in range(1,n):
            for c in range(1,m):
                for b in range(1, min((n-r), (m-c))):
                    if b == 1:
                        max_at[r][c][b] = mat[r][c]
                        maxs = max(maxs, max_at[r][c][b])
                        continue
                    ns = dp[r][c][b-1] + sum(mat[r][c-b+1:c+1]) + sum([mat[rr][c] for rr in range(r-b+1,r+1)])
                    max_at[r][c][b] = max(ns, mat[r][c])
                    maxs = max(maxs, max_at[r][c][b])
        return maxs

```

```python
"""
Print all elements in sorted order from row and column wise sorted matrix
"""

from sys import maxsize

def sortedMatrix(Mat):
    N = len(Mat)
    return [extract(Mat) for _ in range(N*N)]

def youngify(Mat, r, c):
    N = len(Mat)
    vr = Mat[r][c+1] if c+1 < N else maxsize
    vd = Mat[r+1][c] if r+1 < N else maxsize
    if vr == maxsize and vd == maxsize:
        return
    if vr < vd:
        Mat[r][c] = vr
        Mat[r][c+1] = maxsize
        youngify(Mat, r, c+1)
    else:
        Mat[r][c] = vd
        Mat[r+1][c] = maxsize
        youngify(Mat, r+1, c)

def extract(Mat):
    # yield Mat[0][0]
    v = Mat[0][0]
    Mat[0][0] = maxsize
    youngify(Mat, 0, 0)
    return v

def test():
    mat = [[10, 20, 30, 40],
           [15, 25, 35, 45],
           [27, 29, 37, 48],
           [32, 33, 39, 50]]    
    print(sortedMatrix(mat))

test()
```

```python
"""
Create a matrix with alternating rectangles of O and X
e.g. 
Input:  R = 6, C = 7
Output: Following matrix
X X X X X X X
X 0 0 0 0 0 X
X 0 X X X 0 X
X 0 X X X 0 X
X 0 0 0 0 0 X
X X X X X X X 

"""

def create_xo_matrix(R:int, C:int) -> list[list[int]]:
    left, right = 0, C-1
    up, down = 0, R-1
    mat = [[''] * C for _ in range(R)]
    chr = 'X'
    while left <= right and up <= down:
        for c in range(left, right+1):
            mat[up][c] = chr
            mat[down][c] = chr
        for r in range(up, down+1):
            mat[r][left] = chr
            mat[r][right] = chr
        chr = 'O' if chr == 'X' else 'X'
        left += 1
        right -= 1
        up += 1
        down -= 1
    return mat

def test():
    mat = create_xo_matrix(6, 7)
    for r in range(len(mat)):
        print(mat[r])


test()
```
